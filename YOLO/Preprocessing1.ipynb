# Convert Label me Json Annotations to TXT file

```python

import os
from os import walk, getcwd
from PIL import Image
import re

def convert(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)

label_map = {
    "Ani" : 0,
    "Animal":1, 
    "Bin" :2,
    "Building":3,
    "Bus": 4,
    "Car": 5,
    "Graffiti":6,
    "Human":7,
    "Overbridge":8,
    "Pillar":9,
    "Text":10, 
    "Train":11,
    "Truck":12,
    "Underpass":13,
    "Wall":14,
    "wall":15
}
    
    
"""-------------------------------------------------------------------""" 

""" Configure Paths"""   
mypath = "/home/arkil/Desktop/Cmpe257/Graffiti_Detection/Graffiti/new_data/train/"
outpath = "/home/arkil/Desktop/Cmpe257/Graffiti_Detection/Graffiti/text_annotations/"
json_backup ="/home/arkil/Desktop/Cmpe257/Graffiti_Detection/Graffiti/json_backup/"

wd = getcwd()
list_file = open('%s_list.txt'%(wd), 'w')

""" Get input json file list """
json_name_list = []
for file in os.listdir(mypath):
    if file.endswith(".json"):
        json_name_list.append(file)
    

""" Process """
for json_name in json_name_list:
    txt_name = json_name.rstrip(".json") + ".txt"
    """ Open input text files """
    txt_path = mypath + json_name
    print("Input:" + txt_path)
    txt_file = open(txt_path, "r")
    
    """ Open output text files """
    txt_outpath = outpath + txt_name
    print("Output:" + txt_outpath)
    txt_outfile = open(txt_outpath, "a")

    """ Convert the data to YOLO format """ 
    lines = txt_file.read().split('\n')   #for ubuntu, use "\r\n" for windows use "\n"
    for idx, line in enumerate(lines):
        if ("lineColor" in line):
            break 	#skip reading after find lineColor
        if ("label" in line):
            x1 = float(lines[idx+5].rstrip(','))
            print(x1)
            y1 = float(lines[idx+6])
            x2 = float(lines[idx+9].rstrip(','))
            y2 = float(lines[idx+10])
            #cls = line[16:17]
            cls = line[16:30]
            cls = re.findall(r'\b[a-zA-Z]+',cls)
            cls = "".join(cls)
            label_id = label_map[cls]

        #in case when labelling, points are not in the right order
            xmin = min(x1,x2)
            xmax = max(x1,x2)
            ymin = min(y1,y2)
            ymax = max(y1,y2)
            img_path = str('%s/new_data/train/%s.jpg'%(wd, os.path.splitext(json_name)[0]))

            im=Image.open(img_path)
            w= int(im.size[0])
            h= int(im.size[1])

            b = (xmin, xmax, ymin, ymax)
            bb = convert((w,h), b)
            txt_outfile.write(str(label_id) + " " + " ".join([str(a) for a in bb]) + '\n')

    os.rename(txt_path,json_backup+json_name)	#move json file to backup folder

    """ Save those images with bb into list"""
    if(txt_file.read().count("label") != 0):
        list_file.write('%s/new_data/train/%s.jpg\n'%(wd, os.path.splitext(txt_name)[0]))
            
list_file.close()       
```
